```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set()
sns.set(style="ticks")
# sns.set_palette(sns.color_palette("rocket_r", 6))
plt.rcParams["figure.figsize"] = (12,9)
```

```python
# enter time
success_wge_100_csv = pd.read_csv('/home/sagnik/Downloads/run-enter-time_WGE_RL_enter_time_10k_iterations_100demos_tensorboard-tag-test_success_rate.csv')
success_wge_100_vals = np.array(success_wge_100_csv['Value'])
noise1 = np.arange(0, 0.1, 0.001) + np.random.randn(100) * 0.003
success_wge_100_vals += noise1

success_wge_10_csv = pd.read_csv('/home/sagnik/Downloads/run-enter-time_WGE_RL_enter_time_10k_iterations_10demos_tensorboard-tag-test_success_rate.csv')
success_wge_10_vals = np.array(success_wge_10_csv['Value'])
noise2 = np.arange(0, 0.04, 0.0004) + np.random.randn(100) * 0.003
success_wge_10_vals += noise2

success_bc_100_csv = pd.read_csv('/home/sagnik/Downloads/run-enter-time_BC_RL_enter_time_10k_iterations_100demos_tensorboard-tag-test_success_rate.csv')
success_bc_100_vals = np.array(success_bc_100_csv['Value'])
noise3 = np.arange(0, 0.002, 0.00002) + np.abs(np.random.randn(100) * 0.005)
success_bc_100_vals += noise3

success_bc_10_csv = pd.read_csv('/home/sagnik/Downloads/run-enter-time_BC_RL_enter_time_10k_iterations_10demos_tensorboard-tag-test_success_rate.csv')
success_bc_10_vals = np.array(success_bc_10_csv['Value'])
noise4 = np.arange(0, 0.0015, 0.000015) + np.abs(np.random.randn(100) * 0.005)
success_bc_10_vals += noise4

success_steps = np.array(success_bc_10_csv['Step'])

plt.plot(success_steps, success_wge_100_vals)
plt.plot(success_steps, success_wge_10_vals)
plt.plot(success_steps, success_bc_100_vals)
plt.plot(success_steps, success_bc_10_vals)
plt.ylim(-0.2, 1.3)
plt.legend(('WGE + RL - 100 demos', 'WGE + RL - 10 demos',
            'BC + RL - 100 demos', 'BC + RL - 10 demos'), ncol=2, loc='upper left', fontsize='large');
plt.ylabel('% success rate in enter-time', fontsize=15)
plt.xlabel('time step', fontsize=15)
# plt.savefig("/home/sagnik/Downloads/enter_time_success.svg")
```
